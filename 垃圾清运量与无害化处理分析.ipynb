## 垃圾清运量饼图&人口对比

```python
from pyecharts.faker import Faker
from pyecharts import options as opts
from pyecharts.charts import Bar, Page, Pie, Timeline


def timeline_pie() -> Timeline:
    attr = Faker.choose()
    tl = Timeline()
    for i in range(2009, 2018):
        pie = (
            Pie()
            .add(
                "数值",
                list(zip(list(df.地区),list(df["{}".format(i)]))),
                rosetype="radius",
                radius=["30%", "55%"],
            )
            .set_global_opts(title_opts=opts.TitleOpts("垃圾清运量".format(i)))
        )
        tl.add(pie, "{}".format(i))
    return tl
```

```python
timeline_pie().render_notebook()
```

```python
import pyecharts
import pandas as pd
df3 = pd.read_csv('renkou.csv',encoding='gbk')
df3
```

```python

def timeline_bar() -> Timeline:
    x = list(df3.地区)
    tl = Timeline()
    for i in range(2009, 2018):
        bar = (
            Bar()
            .add_xaxis(x)
            .add_yaxis("各省垃圾清运量", list(df["{}".format(i)]))
            .add_yaxis("各省人口", list(df3["{}".format(i)]))
            .reversal_axis()
            .set_global_opts(title_opts=opts.TitleOpts("{}年各省产生垃圾与人口对比".format(i)),
                            datazoom_opts=opts.DataZoomOpts(orient="vertical",range_start="50",range_end="60"))
            .set_series_opts(
            label_opts=opts.LabelOpts(is_show=False),
        )
        )
        tl.add(bar, "{}".format(i))
    return tl

Timeline_bar = timeline_bar()
timeline_bar().render_notebook()
```

##### 数据分析
由上可见，人口的确是垃圾产量的一大原因。人口越多的地方，垃圾清运量越高，如广州，而西藏等人少的地方垃圾清运量很低

# 垃圾清运量地图

```python
def timeline_map() -> Timeline:
    tl = Timeline()
    for i in range(2009, 2018):
        map0 = (
            Map()
            .add(
                "垃圾清运量（单位：万吨）",list(zip(list(df.地区),list(df["{}".format(i)]))),"china",is_map_symbol_show = False
            )
            .set_global_opts(
                title_opts=opts.TitleOpts(title="{}各省垃圾清运量".format(i),subtitle="",
                                          subtitle_textstyle_opts=opts.TextStyleOpts(color="red",font_size=18,font_style="italic")),
                visualmap_opts=opts.VisualMapOpts(min_=16.3,max_=3035.4),
            
            )
        )
        tl.add(map0,"{}".format(i))
    return tl
timeline_map().render_notebook()
```

### 垃圾无害化处理率

```python
import pyecharts
import pandas as pd
df2 = pd.read_csv('wuhai.csv',encoding='gbk')
df2
```

```python
list(df2.地区)
```

```python
def timeline_map() -> Timeline:
    tl = Timeline()
    for i in range(2009, 2018):
        map0 = (
            Map()
            .add(
                "垃圾无害化清理率（单位：百分比）",list(zip(list(df2.地区),list(df2["{}年".format(i)]))),"china",is_map_symbol_show = False
            )
            .set_global_opts(
                title_opts=opts.TitleOpts(title="{}各省垃圾无害化清理率".format(i),subtitle="",
                                          subtitle_textstyle_opts=opts.TextStyleOpts(color="red",font_size=18,font_style="italic")),
                visualmap_opts=opts.VisualMapOpts(min_=29.9,max_=100),
            
            )
        )
        tl.add(map0,"{}年".format(i))
    return tl
timeline_map().render_notebook()
```

* 垃圾无害化处理的工艺主要有:卫生填埋、堆肥和焚烧三种。<br>

由垃圾无害化处理率地图可见，一开始，各省的处理率都十分不平均，高的高达97，低的低达29，西藏在2016年之前甚至没有数据。直到2016年后，全国的垃圾无害化处理都基本达到了90%。

###### 我对影响垃圾无害化处理的因素的猜测如下：
1.高纬度的地方天气寒冷，卫生填埋在地下分解的时间很长，因此导致无害化处理的进度缓慢。<br>
2.西藏等地区，并不进行农业，堆肥的无害化处理方式并不适合此些地区，也将导致其无法无害化处理垃圾<br>
3.人口密度高的地区无法进行焚烧、填埋，因为人类生活用地面积就极大
